# Insertion-sort vs Bucket-sort


# Introduzione
_Insertion Sort_ e _Bucket Sort_ sono due tra i principali __algoritmi di ordinamento__ utilizzati per organizzare una lista di elementi in ordine __crescente__ o __decrescente__. Sebbene entrambi abbiano l'obiettivo di ordinare dati, differiscono significativamente in termini di efficienza e applicabilità in base alla distribuzione e alla dimensione del dataset.

_Insertion Sort_ è un algoritmo __semplice__ ed efficace per dataset di __piccole dimensioni__ o per liste già parzialmente ordinate. Tuttavia, quando applicato a grandi insiemi di dati, può risultare __inefficiente__ a causa della sua __complessità computazionale quadratica__.

D'altra parte, _Bucket Sort_ è progettato per migliorare le prestazioni su insiemi di dati __uniformemente distribuiti__, dividendo i dati in sottogruppi (_bucket_) da ordinare separatamente. Questo approccio permette a _Bucket Sort_ di raggiungere una complessità media __lineare__ in condizioni ottimali, rendendolo una scelta __efficiente__ su grandi dataset con una distribuzione adatta.


# Insertion Sort

# Descrizione
_Insertion Sort_ è un algoritmo di ordinamento semplice che costruisce l'array ordinato un elemento alla volta, confrontando ogni elemento con quelli già ordinati e inserendolo nella posizione corretta. È efficiente per piccoli dataset o liste quasi ordinate, ma diventa inefficiente su dataset di grandi dimensioni, con complessità O(n²).


# Implementazione

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key
    return arr
```

# Esempio di utilizzo


```python
arr = [12, 11, 13, 5, 6, 18, 7, 9, 1, 0]
print("Lista ordinata", insertion_sort(arr))
```

# Tempo di esecuzione
L'_Insertion Sort_ ha una complessità temporale di O(n²) nel caso peggiore, che si verifica quando la lista è ordinata in modo decrescente. In questo scenario, ogni elemento deve essere confrontato con tutti gli altri già ordinati, comportando un numero elevato di confronti e spostamenti. Tuttavia, nel caso migliore, su una lista già ordinata, l'algoritmo ha una complessità di O(n).


# Bucket Sort

# Descrizione
_Bucket Sort_ è un algoritmo di ordinamento che suddivide gli elementi da ordinare in diversi gruppi chiamati __bucket__ (secchi), ordinando poi ogni bucket separatamente tramite un altro algoritmo di ordinamento, come _Insertion Sort_.  
Questo approccio è particolarmente efficace quando gli elementi da ordinare sono distribuiti in modo uniforme su un intervallo, risultando in una complessità media di __O(n)__ per alcuni casi. Tuttavia, la sua efficienza dipende fortemente dalla distribuzione dei dati e dalla scelta dell'algoritmo di ordinamento interno.


# Implementazione

```python
def bucket_sort(arr):
    n = len(arr)
    buckets = [[] for _ in range(n)]
    
    for num in arr:
        # Assicurati che l'indice non superi i limiti
        index = min(int(num * n), n - 1)  # Prevenire errori di indice
        buckets[index].append(num)

    for i in range(n):
        insertion_sort(buckets[i]) 

    return [num for bucket in buckets for num in bucket]
```

# Esempio di utilizzo

```python
arr = [0.897, 0.565, 0.656, 0.1234, 0.665, 0.3434]
print("Lista ordinata", bucket_sort(arr))
```

# Tempo di esecuzione
Per quanto riguarda il _Bucket Sort_, l'algoritmo presenta una complessità temporale media di __O(n)__ quando i dati sono distribuiti __uniformemente__. In questo caso, gli elementi sono suddivisi tra i bucket in modo relativamente __bilanciato__, permettendo un numero ridotto di __confronti__ all'interno di ciascun bucket e riducendo così il tempo di esecuzione complessivo.

Nel caso __peggiore__, invece, il _Bucket Sort_ può richiedere fino a __O(n²)__, specialmente se i dati sono distribuiti in modo __irregolare__. In tale scenario, un numero sproporzionato di elementi può finire in uno o pochi bucket, comportando un numero elevato di confronti e aumentando i tempi di ordinamento.

# Testing

```python
import random
from timeit import default_timer as timer
import matplotlib.pyplot as plt
import numpy as np

# Funzione di normalizzazione
def normalize(arr):
    min_value = min(arr)
    max_value = max(arr)
    return [(num - min_value) / (max_value - min_value) for num in arr]

# Funzioni per generare liste
def listaCasuale(num_elementi):
    return [random.random() for _ in range(num_elementi)]

def listaOrdinata(num_elementi):
    return sorted([random.random() for _ in range(num_elementi)])

def listaOrdinataDecrescente(num_elementi):
    return sorted([random.random() for _ in range(num_elementi)], reverse=True)

# Funzione di test
def test(alg_lista, num_elementi, iterazioni):
    dimensioni_lista = []
    insertion_sort_t = []
    bucket_sort_t = []
    
    lista_max = alg_lista(num_elementi * iterazioni)

    for i in range(iterazioni):
        n = num_elementi * (i + 1)
        dimensioni_lista.append(n)
        print(f"Numero di elementi: {n}")
        
        lista = lista_max[:n]
        copia_lista = lista.copy()
        
        # Normalizza la lista
        copia_lista_normalizzata = normalize(copia_lista)

        # Testa Insertion Sort
        start = timer()
        insertion_sort(lista)
        stop = timer()
        insertion_sort_t.append(stop - start)
        
        # Testa Bucket Sort
        start = timer()
        bucket_sort(copia_lista_normalizzata)
        stop = timer()
        bucket_sort_t.append(stop - start)

    # Crea il grafico che mostra il confronto temporale
    plt.plot(dimensioni_lista, insertion_sort_t, label='Insertion Sort')
    plt.plot(dimensioni_lista, bucket_sort_t, label='Bucket Sort')
    plt.xlabel('Numero di elementi')
    plt.ylabel('Tempo di esecuzione (secondi)')
    plt.title('Confronto tra Insertion Sort e Bucket Sort')
    plt.legend()
    plt.show()

# Funzione principale
def main():
    num_elementi = 2000
    iterazioni = 20
    print('Test con lista casuale...')
    test(listaCasuale, num_elementi, iterazioni)
    print('Test con lista ordinata...')
    test(listaOrdinata, num_elementi, iterazioni)
    print('Test con lista ordinata in ordine decrescente...')
    test(listaOrdinataDecrescente, num_elementi, iterazioni)

main()
```

### Confronto tra Insertion Sort e Bucket Sort

Nei test con **lista casuale**, **Bucket Sort** si dimostra nettamente più efficiente rispetto a **Insertion Sort**. **Bucket Sort** mantiene tempi di esecuzione molto bassi, mentre **Insertion Sort** mostra una crescita esponenziale del tempo con l'aumentare degli elementi. Questo conferma la complessità media lineare di **Bucket Sort**, rispetto alla complessità quadratica di **Insertion Sort**.

Nel test con **lista ordinata**, i risultati sono interessanti: **Bucket Sort** presenta picchi nei tempi di esecuzione a causa della distribuzione non uniforme degli elementi nei bucket. Al contrario, **Insertion Sort** è estremamente efficiente, con tempi quasi costanti, grazie alla sua ottimizzazione per liste già ordinate.

Nel test con **lista ordinata casualmente**, **Bucket Sort** continua a mostrare la sua efficienza costante, mentre **Insertion Sort** soffre nuovamente della crescita esponenziale tipica del suo comportamento su liste non ordinate.

Nel test su **lista ordinata in ordine decrescente**, **Insertion Sort** soffre la sua complessità peggiore, con una crescita esponenziale del tempo di esecuzione all'aumentare del numero di elementi. Al contrario, **Bucket Sort** mantiene prestazioni stabili, con tempi di esecuzione molto più bassi e quasi costanti, confermando ancora una volta la sua efficienza.

### Conclusione

**Bucket Sort** si dimostra ideale per la maggior parte degli scenari, garantendo prestazioni costanti anche con liste disordinate o decrescenti. Tuttavia, **Insertion Sort** può essere più vantaggioso in situazioni specifiche, come con liste già ordinate, dove risulta particolarmente efficiente.
